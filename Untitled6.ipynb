```python
import pandas as pd
df = pd.read_csv('survey lung cancer.csv')
```

```python
import matplotlib.pyplot as plt  # Mengimpor pustaka matplotlib untuk visualisasi data

# Membuat plot
fig, ax = plt.subplots()  # Membuat figure dan axes untuk plot

# Mengelompokkan data berdasarkan jenis kelamin dan membuat scatter plot
for gender, subset in df.groupby('GENDER'):  # Mengelompokkan data berdasarkan kolom 'GENDER'
    ax.scatter(subset['AGE'], subset['SMOKING'], label=gender)  # Membuat plot titik untuk 'AGE' vs 'SMOKING' untuk setiap kelompok gender

# Menambahkan detail plot
plt.legend()  # Menambahkan legenda untuk menunjukkan kelompok berdasarkan jenis kelamin
plt.title('Sebaran Data Berdasarkan Umur, Merokok, dan Jenis Kelamin')  # Menambahkan judul untuk plot
plt.xlabel('Umur (AGE)')  # Menambahkan label untuk sumbu X
plt.ylabel('Merokok (SMOKING)')  # Menambahkan label untuk sumbu Y
plt.grid(True)  # Menambahkan grid untuk memudahkan pembacaan plot
plt.show()  # Menampilkan plot
```

```python
import numpy as np  # Mengimpor library NumPy untuk operasi array
import pandas as pd  # Mengimpor library Pandas untuk manipulasi data
from sklearn.preprocessing import LabelBinarizer  # Mengimpor LabelBinarizer untuk encoding label

# Mendapatkan nama kolom aktual dari DataFrame
actual_columns = df.columns.tolist()  # Mengambil daftar nama kolom dari DataFrame
print(f"Available columns: {actual_columns}")  # Menampilkan nama kolom yang tersedia

# Menyesuaikan nama kolom yang dipilih berdasarkan kolom yang tersedia
features_to_select = ['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
                       'CHRONIC DISEASE', 'WHEEZING', 'ALCOHOL CONSUMING',
                       'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']

# Memeriksa apakah ada fitur yang diinginkan yang hilang
missing_features = [f for f in ['FATIGUE', 'ALLERGY'] if f not in actual_columns]  # Mencari fitur yang hilang
if missing_features:  # Jika ada fitur yang hilang
    print(f"Warning: Features not found in the DataFrame: {missing_features}")  # Menampilkan peringatan

# Menyusun data latih (features) dan target (label)
x_train = np.array(df[features_to_select])  # Mengambil fitur yang dipilih dari DataFrame dan mengubahnya menjadi array NumPy

y_train = np.array(df['LUNG_CANCER'])  # Mengambil kolom target 'LUNG_CANCER' dan mengubahnya menjadi array NumPy

# Menampilkan x_train dan y_train
print(f'x_train:\n{x_train}\n')  # Menampilkan data fitur
print(f'y_train:\n{y_train}\n')  # Menampilkan data target
```

```python
from sklearn.preprocessing import LabelBinarizer  # Mengimpor LabelBinarizer dari pustaka sklearn untuk mengubah label menjadi format biner

# Membuat instance dari LabelBinarizer
lb = LabelBinarizer()

# Melakukan binarisasi pada data target y_train
y_train = lb.fit_transform(y_train)  # Mengubah label target menjadi format biner (0 dan 1)

# Menampilkan hasil binarisasi y_train
print(f'y_train:\n{y_train}\n')  # Menampilkan data target yang telah dibinarisasi
```

```python
# Flatten y_train agar menjadi array satu dimensi
y_train = y_train.flatten()

# Menampilkan hasil flattening y_train
print(f'y_train:\n{y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Menentukan jumlah tetangga terdekat yang digunakan
k = 3

# Membuat dan melatih model K-Nearest Neighbors (KNN)
model = KNeighborsClassifier(n_neighbors=k)
model.fit(x_train, y_train)

```

```python
#misal ini adalah data baru untuk diprediksi
age = 30
smoking = 1  # Anggap 1 = Merokok, 0 = Tidak Merokok
yellow_fingers = 1  # Misalnya memiliki jari kuning
anxiety = 0  # Tidak merasa cemas
peer_pressure = 0  # Tidak terpengaruh tekanan teman
chronic_disease = 0  # Tidak memiliki penyakit kronis
fatigue = 1  # Merasa lelah
allergy = 0  # Tidak memiliki alergi
wheezing = 0  # Tidak memiliki mengi
alcohol_consuming = 1  # Mengkonsumsi alkohol
coughing = 1  # Batuk
shortness_of_breath = 0  # Tidak sesak napas


# Membuat array untuk data baru, menyesuaikan dengan 12 fitur yang diharapkan model
x_new = np.array([age, smoking, yellow_fingers, anxiety, peer_pressure, chronic_disease,
                  fatigue, allergy, wheezing, alcohol_consuming, coughing, shortness_of_breath]).reshape(1, -1)

print(x_new)
```

```python
# Menggunakan model untuk melakukan prediksi berdasarkan data baru
y_new = model.predict(x_new)  # Memanggil metode predict pada model untuk mendapatkan prediksi dari data fitur baru

# Menampilkan hasil prediksi
print(f'Prediksi: {y_new}')  # Menampilkan hasil prediksi yang dihasilkan oleh model
```

```python
# Mengubah hasil prediksi dari format biner kembali ke label aslinya
lb.inverse_transform(y_new)  # Menggunakan metode inverse_transform pada LabelBinarizer untuk mendapatkan label asli dari hasil prediksi biner
```
